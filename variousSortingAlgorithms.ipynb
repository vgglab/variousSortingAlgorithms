```python
class SortingAlgorithms:
  def __init__(self, array):
    self.array = array

  def __isSorted(self, arrayCopy):
    for i in range(len(arrayCopy) - 1):
      if arrayCopy[i] > arrayCopy[i+1]:
        return False
    return True

  def monkeySort(self):
    import random, time
    startTime = time.time()
    arrayCopy = self.array[:]
    while not self.__isSorted(arrayCopy):
      random.shuffle(arrayCopy)
      # print('Shuffled array: ', arrayCopy)
    print('MonkeySort Sorted array: ', arrayCopy, ' taken time: ', f'{time.time() - startTime:.5f}')

  def sleepSort(self):
    import time
    startTime = time.time()
    arrayCopy = self.array[:]
    sortedArray = list()
    while len(arrayCopy) > len(sortedArray):
      time.sleep(1)
      elapsed = int(time.time() - startTime)
      if elapsed in arrayCopy:
        sortedArray.append(elapsed)
        # print('Intermediate array: ', sortedArray)
    print('SleepSort Sorted array: ', sortedArray, ' taken time: ', f'{time.time() - startTime:.5f}')


  def bubbleSort(self):
    arrayCopy = self.array[:]
    isAlreadySorted = False
    import time
    startTime = time.time()
    for i in range(len(arrayCopy)- 1):
      if isAlreadySorted is True:
        print('BubbleSort', ' taken time: ', f'{time.time() - startTime:.5f}')
        break
      else:
        # print('Intermediate array: ', arrayCopy)
        for j in range(len(arrayCopy) - 1 - i):
          if arrayCopy[j] > arrayCopy[j+1]:
            arrayCopy[j], arrayCopy[j+1] = arrayCopy[j+1], arrayCopy[j]
            isAlreadySorted = False
          else:
            isAlreadySorted = True
            continue

  def selectionSort(self):
    import time
    startTime = time.time()
    arrayCopy = self.array[:]
    for i in range(len(arrayCopy) - 1):
      minimum = i
      # print('Intermediate array: ', arrayCopy)
      for j in range(i+1, len(arrayCopy)):
        if arrayCopy[minimum] < arrayCopy[j]:
          continue
        else:
          arrayCopy[minimum], arrayCopy[j] = arrayCopy[j], arrayCopy[minimum]
    print('SelectionSort', ' taken time: ', f'{time.time() - startTime:.5f}')

  def __mergeSortedArray(self, array1, array2):
    i = j = 0
    sortedArray = list()
    while i < len(array1) and j < len(array2):
      if array1[i] < array2[j]:
        sortedArray.append(array1[i])
        i+=1
      else:
        sortedArray.append(array2[j])
        j+=1
    while i < len(array1):
      sortedArray.append(array1[i])
      i+=1
    while j < len(array2):
      sortedArray.append(array2[j])
      j+=1
    return sortedArray


  def __recursiveMergeSort(self, array):
    if len(array) == 1:
      return array

    leftArray = array[:len(array)//2]
    rightArray = array[len(array)//2:]

    leftArray = self.__recursiveMergeSort(leftArray)
    rightArray = self.__recursiveMergeSort(rightArray)

    return self.__mergeSortedArray(leftArray, rightArray)

  def mergeSort(self):
    import time
    arrayCopy = self.array[:]
    startTime = time.time()
    sortedArray = self.__recursiveMergeSort(arrayCopy)
    stopTime = time.time()
    print('MergeSort', ' taken time: ', f'{stopTime - startTime:.5f}')
```
